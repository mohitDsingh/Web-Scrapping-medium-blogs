### How to scratch a Web-Page which loads more material on *Scrolling*

```
Using .scrollTo() we can physically scroll the web page. When using selenium web driver we can save the web page to perform operations. Also we can directly use BeautifulSoup to scrape the data.

BeautifulSoup is a Python library for pulling data out of HTML and XML files.

Using inspect element of web page we can find out the container that contains the desired data. For this right click anywhere on web page and and select inspect or ctrl+shift+i.
```

Now lets make Extracting Blogs for any topic for more than 100 Blog's automatic by make it into a fuction

````python
def get_topic_page(topic):
    #Construct the URL
    blog_url = 'https://medium.com/tag/' + topic.replace(" ", "-")
    
    #Get the HTML Page content using requests
    driver = webdriver.Chrome(chromedriver_autoinstaller.install())
    driver.get(blog_url)

    ScrollNumber = 50
    for i in range(1,ScrollNumber):
        driver.execute_script("window.scrollTo(1,50000)")
        time.sleep(5)
    

    #Construct a beautiful_soup Document
    doc = BeautifulSoup(driver.page_source)
    
    return doc
````

````python
#Extracting Trendicg Blogs For ML
ml = get_topic_page("machine learning")
````

````python
# Extracting articles for ml
ml_articles = ml.find_all('article')
print(len(ml_articles))
ml_blog_title = get_blog_title(ml_articles)
len(ml_blog_title)
````

````python
def parse_blog_detail(article_tag):
    for article in article_tag:
        #Blog_title
        blog_titles = get_blog_title(article)[0]
        #Owner's Username
        username = get_usernames(article)[0]
        #blog URl
        blog_url = get_blog_link(article)[0]
        #Owners Profile link
        profile_link = get_user_links(article)[0]
        #Time to Read
        read_time = get_read_time(article)[0]
        #Uplaod Tag's
        upload_tag = get_upload_tags(article)[0]
        #Return a Dictionory
        return {
            'Blog Titile' : blog_titles,
            'Blog Link' : blog_url,
            'Owner Name' : username,
            'Owner Profile' : profile_link,
            'Read Time' : read_time,
            'Upload Tag' : upload_tag
    }
````

````python
ml_trending_blog_details = parse_blog_detail(ml_articles)
ml_trending_blog_details
````

````python
#Extracting for tag "Machine-Learning"
ml_trending_blog_details = [parse_blog_detail(article) for article in ml_articles]
print(len(ml_trending_blog_details))
ml_trending_blog_details
````

````python
def get_trending_blogs(doc):
    article_tags = doc.find_all('article')
    trending_blogs = [parse_blog_detail(tag) for tag in article_tags]
    return trending_blogs
````

````python
#Extracting for tag "Data Analysis"
data_analysis_trending = get_trending_blogs(get_topic_page("data analysis"))
data_analysis_trending[:5]
````
